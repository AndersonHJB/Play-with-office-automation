```python
import xlsxwriter
import random
from faker import Faker
import datetime
wb = xlsxwriter.Workbook("一万个好员工信息.xlsx")
sheet = wb.add_worksheet('Sheet1')
```

```python
headings = ['name','gender','join-year','join-month','department','job','job-name','company']
sheet.write_row('A1',headings)
```

```python
department_info = {
    '财务':['出纳','会计'],
    '开发':['平台开发','后端开发','前端开发'],
    '测试':['黑盒测试','白盒测试'],
    '运维':['数据库运维','服务器管理'],
    '运营':['新媒体运营','传统运营'],
    '设计':['原画','插画'],
    '管理':['公司领导','技术领导'],
}
```

```python
fake = Faker("zh_CN")
for i in range(2,10002):
    name = fake.name()
    gender = '男' if fake.boolean() else '女'
    
    join_year = random.choice(range(2001,2020))
    join_month = random.choice(range(1,13))
    
    department = random.choice(list(department_info.keys()))
    job = random.choice(department_info[department])
    job_name = random.choice(['初级','中级','高级','特聘']) + job
    
    company = fake.company()
    
    #today_datetime = datetime.datetime.now()
    #year = today_datetime.year
    #month = today_datetime.month
    #day = today_datetime.day
    
    line_data= [name,gender,join_year,join_month,department,job,job_name,company]
    #line_data= [name,gender,join_year,join_month,department,job,job_name,company, year,month,day]
    print(line_data)
    sheet.write_row('A{}'.format(i), line_data)
```

```python
wb.close()
```
