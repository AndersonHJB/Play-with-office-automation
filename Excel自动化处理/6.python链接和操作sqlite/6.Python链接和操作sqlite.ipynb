```python
import sqlite3
```

```python
firstdb = sqlite3.connect('first.db')
```

```python
# 查询语句
query_sql = "select * from info"
for result in firstdb.execute(query_sql):
    print(result)
```

```python
# 删除特定数据
delete_sql = "delete from info where id = 3 "

firstdb.execute(delete_sql)
firstdb.commit()

# 查询并输出
query_sql = "select * from info"
for result in firstdb.execute(query_sql):
    print(result)
```

```python
# 插入数据
insert_sql = "insert into info(title, content, author) values ('第{}个标题', '随机的第{}个内容', '匿名')"
for i in range(100,120):
    sql = insert_sql.format(i,i*2)
    firstdb.execute(sql)
    firstdb.commit()

    
# 查询并输出
query_sql = "select * from info"
for result in firstdb.execute(query_sql):
    print(result)
```

```python
# 更新数据
update_sql = "update info set author = '小布' where id >= 10"
firstdb.execute(update_sql)
firstdb.commit() # execute执行了代码，必须commit才可以

# 查询并输出
query_sql = "select * from info"
for result in firstdb.execute(query_sql):
    print(result)
```
