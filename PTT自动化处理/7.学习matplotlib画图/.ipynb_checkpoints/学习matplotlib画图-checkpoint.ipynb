```python
from matplotlib import pyplot as plt
import pandas as pd

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


df = pd.read_excel('ten-thousand.xlsx')
```

```python
pd.pivot_table(df, index=['company'], values=['name'], aggfunc='count')
```

```python
company_person = pd.pivot_table(df, index=['company'], values=['name'], aggfunc='count')
company_person.plot.line()
company_person.plot.bar()

plt.show()
```

```python
company_person = pd.pivot_table(df, index=['company'], values=['name'], aggfunc='count')[:10]
company_person.plot.line()
company_person.plot.bar()

plt.show()
```

```python
gender_person = pd.pivot_table(df, index=['gender'], values=['name'], aggfunc='count')
gender_person.plot.pie(subplots=True)
gender_person.plot.barh()

plt.show()
```
