```python
import openpyxl, math
wb = openpyxl.load_workbook('ten-thousand.xlsx')
sheet = wb.get_sheet_by_name('Sheet1')
```

```python
rows = list(sheet.rows)
one = [cell.value for cell in rows[0]]
for row in rows[1:]:
    data = [str(cell.value) for cell in row]
    print(data)
```

```python
rows = list(sheet.rows)
one = [cell.value for cell in rows[0]]
department_datas = {}
join_year_data = {}
join_month_data = {}
all_male = 0
all_female = 0
for row in rows[1:]:
    name,gender, join_year, join_month, department, job, job_name, company = [str(cell.value) for cell in row]
    join_year_data[join_year] = join_year_data.get(join_year,0) +1
    join_month_data[join_month] = join_month_data.get(join_month,0) +1
    department_datas[department] = department_datas.get(department,{})
    if gender == "男":
        all_male += 1
        department_datas[department]['male'] = department_datas[department].get('male',0) + 1
    else:
        all_female += 1
        department_datas[department]['female'] = department_datas[department].get('female',0) + 1
```

```python
for data in department_datas:
    print(data, department_datas[data])
print(join_month_data)
print(join_year_data)
print(all_male,all_female)
```

```python
max_gcd = math.gcd(all_male,all_female)
malevfemale = "{}:{}".format(all_male//max_gcd, all_female//max_gcd)

max_join_year_count = max(join_year_data.values()) 
max_join_year = " ".join([k for k,v in join_year_data.items() if v==max_join_year_count])


max_join_month_count = max(join_month_data.values()) 
max_join_month = " ".join([k for k,v in join_month_data.items() if v==max_join_month_count])


for department in department_datas.keys():
    department_datas[department]['all'] = department_datas[department]['male'] + department_datas[department]['female']
max_department_count = max([department['all'] for department in department_datas.values()])
max_department =  " ".join([k for k,v in department_datas.items() if v['all']==max_department_count])

print(malevfemale, max_join_year, max_join_month, max_department)


change_dict = {
    'malevfemale':malevfemale,
    'max-join-year':max_join_year,
    'max-join-month':max_join_month,
    'max-department':max_department,
}
print(change_dict)
```

```python
from pptx import Presentation
from pptx.util import Inches
from pptx.shapes.graphfrm import GraphicFrame


prs = Presentation('【模板】好员工信息分析统计.pptx')

slides = prs.slides

for slide in slides:
    for shape in slide.shapes:
        if type(shape) != GraphicFrame:
            for k,v in change_dict.items():
                shape.text = shape.text.replace(k,v)
            print(shape.text)
```

```python
slide_table = slides[2]
shapes = slide_table.shapes

GF = [shape for shape in shapes if type(shape) == GraphicFrame][0]
table = GF.table
rows = table.rows

for row in list(rows)[1:]:
    one,two,three = row.cells
    for data in department_datas.keys():
        if one.text == data:
            all_count = department_datas[data]['all']
            malevfemale = "{}:{}".format(department_datas[data]['male'],department_datas[data]['female'])
            two.text = str(all_count)
            three.text = malevfemale
            
    print(one.text,two.text,three.text)
```

```python
prs.save("New-好员工信息分析统计.pptx")
```
