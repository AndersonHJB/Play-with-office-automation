```python
import pandas as pd
df = pd.read_excel('ten-thousand.xlsx')
```

```python
df.head()
```

```python
# 查看每个公司的总员工人数
pd.pivot_table(df,index=['company'],values=['name'] , aggfunc='count')
```

```python
# 查看每个公司的男女员工人数
pd.pivot_table(df,index=['company','gender'], values=['name'], aggfunc='count')
```

```python
# 查看数据中每年的人数
pd.pivot_table(df,index=['join-year'], values=['name'], aggfunc='count')
```

```python
# 查看数据中每年中每个月的总人数
pd.pivot_table(df,index=['join-year','join-month'], values=['name'], aggfunc='count')
```

```python
# 查看数据中公司各部门人数
pd.pivot_table(df,index=['company','department'], values=['name'], aggfunc='count')
```

```python
# 查看数据中公司部门平均人数
department_person = pd.pivot_table(df,index=['company','department'], values=['name'], aggfunc='count')
pd.pivot_table(department_person, index=['company'], aggfunc='mean')
```
